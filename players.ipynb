```python
news_src = ['https://www.oneesports.gg/league-of-legends/lck-summer-split-2022-roster/','https://www.oneesports.gg/league-of-legends/lpl-summer-split-2022-roster/','https://www.oneesports.gg/league-of-legends/lec-summer-split-2022-roster/','https://www.oneesports.gg/league-of-legends/lcs-summer-split-2022-rosters/']
SummonerName_list = []
Role_list = []
MostPlay_list = []
Country_list = []
Team_list = []
League_list = []
trigg = 0
for i in news_src:
  NameForLink = requests.get(i)
  soup1 = bs4.BeautifulSoup(NameForLink.text, 'lxml')
  N = soup1.find_all('figure', {'class':'wp-block-table'})
  def h_name(x):
    data_list = []
    name_list = []
    for i in x.find_all('td'):
      data_list.append(i.text)
    del data_list[1::2]
    for j in data_list[1::]:
      result = re.search('“(.*)”', j)
      name_list.append(result.group(1))
    return  name_list
  Name_p = list(map(h_name, N))
  def Most_Played(y):
    return y['title']
  for i in Name_p:
    for j in i:
      try:
        src = requests.get('https://www.trackingthepros.com/player/' + j)
        soup2 = bs4.BeautifulSoup(src.text, 'lxml')
        SummonerName = soup2.find('h1')
        SN = SummonerName.text.strip().replace(SummonerName.a.text, "")
        SummonerName_list.append(SN)
        Role = soup2.find('tbody').find_all('td')[3].contents[0]
        Role_list.append(Role)
        try:
          Champ = soup2.find('tbody').find_all('td')[5].contents
          Champs = list(map(Most_Played, Champ))
          MostPlay_list.append(Champs)
        except:
          MostPlay_list.append('Unknown Data')
          trigg = trigg + 1
        if trigg == 1:
          Country = soup2.find('tbody').find_all('td')[7].contents[0]
          Country_list.append(Country)
          League = soup2.find('tbody').find_all('td')[9].contents[0]
          League_list.append(League)
          trigg = trigg - 1
        else:
          Country = soup2.find('tbody').find_all('td')[9].contents[0]
          Country_list.append(Country)
          League = soup2.find('tbody').find_all('td')[11].contents[0]
          League_list.append(League)
        Team = soup2.find('tbody').find_all('a')[0].text
        Team_list.append(Team)
      except:
        pass
#TableForPlayer = pd.DataFrame([SummonerName_list,Role_list,MostPlay_list_1,Country_list,Team_list,League_list]).transpose()
#TableForTeamName = pd.DataFrame([TeamName_list,TeamLeague_list]).transpose()

    
```

```python
def get_unique_team_name (Team_list):
    list_of_unique_team_name = []
    unique_team_name = set(Team_list)
    for i in unique_team_name:
        list_of_unique_team_name.append(i)
    return list_of_unique_team_name
print(get_unique_team_name(Team_list))
```

```python
def get_unique_league (League_list):
    list_of_unique_league = []
    unique_league = set(League_list)
    for i in unique_league:
        list_of_unique_league.append(i)
    return list_of_unique_league
print(get_unique_league(League_list))
```
